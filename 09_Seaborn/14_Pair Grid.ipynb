```python
import numpy as np 
import pandas as pd
import matplotlib. pyplot as plt
import seaborn as sns

%matplotlib inline 
%reload_ext autoreload
%autoreload 2
```

```python
iris = sns.load_dataset("iris")

g = sns.PairGrid(iris, hue="species")
g.map_diag(sns.histplot)       
g.map_offdiag(sns.scatterplot) 
g.add_legend()
plt.show()
```
