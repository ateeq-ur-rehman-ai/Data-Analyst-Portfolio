# Exploratory Data Analysis (EDA) :

## Ploting a bar char and number of counts by Gender:

```python
gender = sns.countplot(x='Gender', hue='Gender', data=df, palette=["blue", "orange"], legend=False)
for bars in gender.containers:
    gender.bar_label(bars)
plt.title("Counts By Gender:",fontsize=18)
plt.show()
```

## Ploting a bar chart for total amount by Gender

```python
sales_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
s_gen=sns.barplot(x='Gender', y='Amount',hue='Gender', data=sales_gender, palette=['blue', 'orange'])
plt.title("Total Sales By Gender:",fontsize=18)
plt.show()
```

*Both graphs show that most buyers are women, and they also spend more money than men.*

## Ploting a bar char and number of counts by Age :

```python
age = sns.countplot(x='Age Group', hue='Gender', data=df, palette=["blue", "orange"], legend=False)
for bars in age.containers:
    age.bar_label(bars)
plt.title("Counts By Age:",fontsize=18)
plt.show()
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
colors = ["blue", "orange", "green", "red", "purple", "brown", "pink"] 
s_age=sns.barplot(x='Age Group', y='Amount',hue='Amount',dodge=False,data=sales_age,palette=colors)
plt.title("Total Sales By Age:",fontsize=18)
plt.show()
```

## Total number of order from top 10 states :

```python
order_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
colors = ["blue", "orange"] * 5 
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data=order_state, x='State', y='Orders',hue='State',dodge=False,palette=colors,legend=False)
plt.title("Order from top 10 states:",fontsize=18)
plt.show()
```

## Total amount/sales form to 10 States :

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
colors = ["blue", "orange"] * 5 
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data=sales_state, x='State', y='Amount',hue='State',dodge=False,palette=colors,legend=False)
plt.title("Sales from top 10 states:",fontsize=18)
plt.show()
```

## Marital Status :

```python
sales_marital = sns.countplot(data = df, x = 'Marital_Status',hue='Marital_Status',palette=["blue", "orange"] ,legend=False )

sns.set(rc={'figure.figsize':(7,5)})
for bars in sales_marital.containers:
    sales_marital.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender',palette=["blue", "orange"],legend=False )
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power.*

## Occupation :

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation',color="orange")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount',color="blue")
```

## Product Category :

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category',color='orange')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount',color='blue')
```
